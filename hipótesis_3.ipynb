# Justificación en inversión de marketing o  merchandising.
•	Hipótesis:
Producciones con una calificación superior a 8 en IMDb suelen tener mayor potencial de convertirse en proyectos rentables, justificando una inversión adicional en marketing o merchandising.

•	Objetivo:
Identificar contenido con calificaciones mayores a 8

•	Análisis:
Filtrar contenido con valores mayores a 8 en los campos imdb_score, asegurando que haya votos registrados en imdb_votes.

```python
import pandas as pd
import pymongo
```

```python
# Datos en formato CSV
data_netflix = pd.read_csv("data_netflix.csv")
mejores_shows = pd.read_csv("Mejores_shows_netflix.csv")
mejores_peliculas = pd.read_csv("Mejores_películas_Netflix.csv")
actores = pd.read_csv("Actores.csv")
```

```python
# Conexión a MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")

# Acceso a la BD
db = client["netflix"]

# Acceso a la colección
collection = db['data_netflix']
```

```python
pipeline = [
   {
    $match: {
      imdb_score: { $gt: 8 },  # Filtro de las producciones con calificación superior a 8
      imdb_votes: { $gt: 0 }   # Asegurando que haya votos registrados
    }
   }
]

# Ejecutar la agregación
result = collection.aggregate(pipeline)
```
