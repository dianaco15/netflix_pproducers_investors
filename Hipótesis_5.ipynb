```python
import pandas as pd
import pymongo
```

```python
# Datos en formato CSV
data_netflix = pd.read_csv("data_netflix.csv")
mejores_shows = pd.read_csv("Mejores_shows_netflix.csv")
mejores_peliculas = pd.read_csv("Mejores_películas_Netflix.csv")
actores = pd.read_csv("Actores.csv")
```

```python
# Conexión a MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")

# Acceso a la BD
db = client["netflix"]

# Acceso a la colección
collection = db['cruce_data_netflix_actores']
```

```python
pipeline = [
   {
    $group: {
      _id: { type: "$type", id: "$id" }, # Agrupación por "type" e "id"
      personal_pelicula: { $sum: 1 }        # Conteo del número de documentos (/direc)
    }
  },
  {
    $project: {
      _id: 0,
      type: "$_id.type",       # Extracción "type" del grupo
      id: "$_id.id",           # Extracción "id" del grupo
      personal_pelicula: 1        # el conteo de actores
    }
  },
  {
    $sort: { personal_pelicula: -1 } # Orden de mayor a menor
  },
	{
	"$out": 'personal_type'
	}
]


result = collection.aggregate(pipeline)

```
