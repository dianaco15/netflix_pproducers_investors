```python
import json
import pandas as pd
```

```python
# abrir el json genero_conteo
with open('genero_conteo.json') as f:
    genero_conteo = json.load(f)

# Convertir el JSON a un DataFrame
df = pd.DataFrame(genero_conteo)

# Mostrar el DataFrame
df


```

```python
# Calcular la correlación entre las columnas
correlation = df[['conteo_shows', 'conteo_peliculas']].corr()

# Mostrar la correlación
print(correlation)
```
