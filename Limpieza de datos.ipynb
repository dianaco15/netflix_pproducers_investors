# LIMPIEZA DE DATA FRAMES

```python
import pandas as pd
```

```python
data_netflix = pd.read_csv('data_netflix.csv')
mejores_shows_netflix = pd.read_csv('mejores Shows Netflix.csv')
mejores_peliculas_netflix = pd.read_csv('mejores peliculas Netflix.csv')

#print(data_netflix.head())
#print(mejores_shows_netflix.head())
print(mejores_peliculas_netflix.head())
```

```python
#Evaluación de datos nulos
mejores_peliculas_netflix.isnull().sum()
```

```python
#Renombración
data_netflix = data_netflix.rename(columns={
    'Id': 'id',
    'Title': 'title',
    'Type': 'type',
    'release_year': 'release_year',
    'age_certification': 'age_certification',
    'Runtime': 'runtime',
    'Genres': 'genres',
    'production_countries': 'production_countries',
    'Seasons': 'seasons',
    'imdb_id': 'imdb_id',
    'imdb_score': 'imdb_score',
    'imdb_votes': 'imdb_votes'
})


mejores_shows_netflix = mejores_shows_netflix.rename(columns={
    'TITLE': 'title',
    'RELEASE_YEAR': 'release_year',
    'SCORE': 'score',
    'NUMBER_OF_VOTES': 'votes',
    'DURATION': 'duration',
    'NUMBER_OF_SEASONS': 'seasons',
    'MAIN_GENRE': 'genre',
    'MAIN_PRODUCTION': 'production_country'
})

mejores_peliculas_netflix = mejores_peliculas_netflix.rename(columns={
    'TITLE': 'title',
    'RELEASE_YEAR': 'release_year',
    'SCORE': 'score',
    'NUMBER_OF_VOTES': 'votes',
    'DURATION': 'duration',
    'MAIN_GENRE': 'genre',
    'MAIN_PRODUCTION': 'production_country'
})
```

```python
# Aseguración de que 'genres' y 'production_countries' son listas
data_netflix['genres'] = data_netflix['genres'].apply(lambda x: eval(x) if isinstance(x, str) else x)
data_netflix['production_countries'] = data_netflix['production_countries'].apply(lambda x: eval(x) if isinstance(x, str) else x)

# Expanción de la columna 'genres' en filas separadas
data_netflix_expanded_genres = data_netflix.explode('genres').reset_index(drop=True)

# Expanción de la columna 'production_countries' en filas separadas
data_netflix_expanded_genres_countries = data_netflix_expanded_genres.explode('production_countries').reset_index(drop=True)

# Verificación
print(data_netflix_expanded_genres_countries)
```

```python
# Eliminación de filas con valores nulos
data_netflix_expanded_genres_countries = data_netflix_expanded_genres_countries.dropna(subset=['imdb_id', 'imdb_score', 'imdb_votes', 'title', 'production_countries', 'genres' ])
```

```python
#Agrupación por 'genres' y calculo de la moda de 'age_certification'
mode_age_certification_by_genre = data_netflix_expanded_genres_countries.groupby('genres')['age_certification'].apply(lambda x: x.mode()[0] if not x.mode().empty else None)

# Remplazo de valores nulos en 'age_certification'
data_netflix_expanded_genres_countries['age_certification'] = data_netflix_expanded_genres_countries.apply(
    lambda row: mode_age_certification_by_genre.get(row['genres'], row['age_certification'])
    if pd.isnull(row['age_certification']) else row['age_certification'], axis=1
)


print(data_netflix_expanded_genres_countries)
```

```python
data_netflix_expanded_genres_countries.to_csv("data_netflix_clean.csv", index=False)

print("El archivo ha sido guardado como 'data_netflix_clean.csv'.")
```

```python
mejores_shows_netflix.to_csv("mejores_shows_netflix.csv", index=False)

print("El archivo ha sido guardado como 'mejores_shows_netflix.csv'.")
```

```python
mejores_peliculas_netflix.to_csv("mejores_peliculas_netflix.csv", index=False)

print("El archivo ha sido guardado como 'mejores_peliculas_netflix.csv'.")
```
